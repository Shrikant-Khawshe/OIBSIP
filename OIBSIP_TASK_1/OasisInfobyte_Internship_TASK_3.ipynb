```python
x = df.drop(axis = 1, columns = ["price","fuelsystem","boreratio","aspiration","symboling","carbody","drivewheel",'enginelocation','enginetype','fueltype','doornumber','cylindernumber'])
y = df["price"]
y
```

#Encoding


```python
temp=df[["fueltype","doornumber","cylindernumber"]]
```

```python
oe = OrdinalEncoder(categories=[['gas', 'diesel'],['four', 'two'],['four', 'eight', 'six', 'five', 'two', 'three', 'twelve']])
```

```python
oe.fit(temp)
```

```python
temp1 = oe.transform(temp)
temp1
```

```python
x = np.concatenate((x,temp1),axis=1)
x
```

# Model Training


```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,random_state=18, test_size=0.20)
```

```python
model=regression.fit(X_train,y_train)
```

```python
y_pred=regression.predict(X_test)
y_pred
```

```python
model.score(X_test,y_test)
```
